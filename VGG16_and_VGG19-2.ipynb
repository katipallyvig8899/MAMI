```python
param_space = {
    'first_convlayer_num_f_filters':[8,16],
    'convlayer_filter_size':[(3,3),(5,5)],
    'sec_convlayer_num_f_filters':[16,32],
    'padding':['same','valid'],
    'maxpool_Size':[(2,2),(4,4)],
    'dense_layer_neuron':[32,64],
    'optimizers':['Adam','rmsprop'],
    'batch_size':[500,1000],
    'dropout_fraction':[0.2,0.3],
    'epochs': [5],
    'activation_func':['relu']
}
```

```python
from tensorflow.keras.layers import *
from tensorflow.keras import Sequential
from keras.callbacks import EarlyStopping

def myconvmodel(X_train,Y_train,X_test,Y_test,params):
    #print(params['convlayer_filter_size'])
    model = Sequential() 
    model.add(Conv2D(params['first_convlayer_num_f_filters'],params['convlayer_filter_size'],input_shape = (120,120,3),activation = 'relu',padding = params['padding'])) # layer1
    model.add(MaxPooling2D(pool_size = params['maxpool_Size'])) # layer2
    model.add(Conv2D(params['sec_convlayer_num_f_filters'],params['convlayer_filter_size'],activation = 'relu',padding = params['padding'])) # layer1
    model.add(MaxPooling2D(pool_size = params['maxpool_Size'])) # layer4

    model.add(Flatten()) #layer5
    model.add(Dense(params['dense_layer_neuron'],activation = params['activation_func']))# layer6
    model.add(Dropout(params['dropout_fraction']))
    model.add(Dense(2,activation = 'softmax')) # layer7

    model.compile(optimizer = params['optimizers'],loss='binary_crossentropy',metrics = ['acc']) 
    es = EarlyStopping(monitor = 'val_loss',mode = 'min',patience = 1)
    history = model.fit(x = X_train_data,y = Y_train,batch_size =params['batch_size'], epochs = 5,callbacks = [es],validation_split=0.2)

    return history,model
```
