```python
from matplotlib import pyplot as plt
```

```python
def preprocess_image(im):
    im= cv2.bilateralFilter(im,5,55,60)
    im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    _, im = cv2.threshold(im, 127,255,cv2.THRESH_BINARY)
    plt.figure(figsize=(10,10))
    plt.imshow(im, cmap='gray'); plt.xticks([]); plt.yticks([])
    return im
def preprocess_text(text):
    user_name = '@[A-Za-z0-9]+' # Pattern for matching the user names in the tweet
    has_tag = '#[A-Za-z0-9]+' # Pattern for matching the hasgtags in the tweer
    url = 'https?:\/\/[A-Za-z0-9\.\/\-]+' # Pattern for matching the URLs in the tweet
    text = text.lower()  # Lower casing all the characters
    text = re.sub(r'[^\w\s]','',text)#cleaning
    text = re.sub(user_name, '', text) # Replace the username with an empty characrer
    text = re.sub(has_tag,'',text) # Replace the hashtags with an empty characrer
    text = re.sub(url, '', text) # Replace the URLs with an empty characrer
    text = re.sub("[^@a-zA-Z0-9]", " ",text) # Replace all the characters except alphabets and digits from the tweet
    text = re.sub("rt\s+",'',text) # Replace the retweet character "rt" from the tweet
    return text
bilateral_bin=[]
bilateral_bin_img=[]
for j in cv_img:
    i=preprocess_image(j)
    bilateral_bin_img.append(i)
    text = pt.image_to_string(i,lang='eng')
    text=text.replace('\n',' ')
    text=preprocess_text(text)
    bilateral_bin.append(text)
print(bilateral_bin)
#Text_info=pd.DataFrame(list(zip(img2,text_img)),columns=['IMAGEID','IMAGE_TEXT'])
#Text_info.to_excel(r'MAMI_TEXT_TRAIL_bilateral_Method.xlsx', index = True, header=True)
```

```python
def preprocess_image1(im):
    im = cv2.blur(im,(5,5))
    im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    _, im = cv2.threshold(im, 240, 255, 1)
    plt.figure(figsize=(10,10))
    plt.imshow(im, cmap='gray'); plt.xticks([]); plt.yticks([])
    return im
average_bin=[]
for j in cv_img:
    i=preprocess_image1(j)
    text = pt.image_to_string(i,lang='eng')
    text=text.replace('\n',' ')
    text=preprocess_text(text)
    average_bin.append(text)
print(average_bin)
#Text_info=pd.DataFrame(list(zip(img2,text_img)),columns=['IMAGEID','IMAGE_TEXT'])
#Text_info.to_excel(r'MAMI_TEXT_TRAIL_average_Method.xlsx', index = True, header=True)
```

```python
def preprocess_image2(im):
    gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (3,3), 0)
    _,im = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    #plt.figure(figsize=(10,10))
    #plt.imshow(im, cmap='gray'); plt.xticks([]); plt.yticks([])
    # Morph open to remove noise and invert image
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
    opening = cv2.morphologyEx(im, cv2.MORPH_OPEN, kernel, iterations=1)
    invert = 255 - opening
    return invert
median_bin=[]
for j in cv_img:
    i=preprocess_image2(j)
    text = pt.image_to_string(i,lang='eng')
    text=text.replace('\n',' ')
    text=preprocess_text(text)
    median_bin.append(text)
print(median_bin)
Text_info=pd.DataFrame(list(zip(img2,median_bin)),columns=['IMAGEID','IMAGE_TEXT'])
Text_info.to_excel(r'MAMI_TEXT_TRAIL_median_Method.xlsx', index = True, header=True)
```
